```python
import pandas as pd
import requests
```

```python
df = pd.DataFrame()

for i in range(1, 10):
    response = requests.get('https://openlibrary.org/subjects/romance.json?limit=5'.format(i))
    temp_df = pd.DataFrame(response.json()['works']).head(5)[['title','edition_count','subject','ia_collection','printdisabled','availability']]
    df = pd.concat([df, temp_df], ignore_index=True)
```

```python
# by writing this piece of code the above data frame to csv
df.to_csv('Books.csv')
```
